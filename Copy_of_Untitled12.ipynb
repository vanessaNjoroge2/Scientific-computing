```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.distributions.empirical_distribution import ECDF

# Load the Puromycin dataset (manually creating similar structure)
# In R, it has columns: 'rate' (reaction rate), 'state' (Treated/Untreated)
data = pd.read_csv("https://vincentarelbundock.github.io/Rdatasets/csv/datasets/Puromycin.csv")

# Separate treated and untreated groups
treated = data[data['state'] == 'treated']['rate']
untreated = data[data['state'] == 'untreated']['rate']

# Compute ECDFs
ecdf_treated = ECDF(treated)
ecdf_untreated = ECDF(untreated)

# Plot ECDFs
plt.step(ecdf_treated.x, ecdf_treated.y, where='post', label='Treated', color='black')
plt.step(ecdf_untreated.x, ecdf_untreated.y, where='post', label='Untreated', color='blue')

# Plot settings
plt.title("Treated versus Untreated - ECDF")
plt.xlabel("Reaction Rate")
plt.ylabel("ECDF")
plt.xlim(60, 200)
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # for 3D plotting

# Create x and y sequences
x = np.linspace(-8, 8, 100)
y = np.linspace(-8, 8, 100)
x, y = np.meshgrid(x, y)

# Define the function
r = np.sqrt(x**2 + y**2)
z = np.sin(r) / r
z[np.isnan(z)] = 1  # Handle the divide-by-zero at r=0

# Plot using wireframe
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_wireframe(x, y, z, rstride=2, cstride=2)

# Optional: remove axes and box (like in R)
ax.set_axis_off()

plt.show()
```
